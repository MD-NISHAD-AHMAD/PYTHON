## FREQUENCY COUNTING.

```python
string = " Hard work is worthless for those that don't believe in themselves"
```

```python
d = {}
for char in string:
    d[char] = d.get(char, 0) + 1
```

```python
d
```

```python
string = " Hard work is worthless for those that don't believe in themselves"

word_frequency = {}
character_frequency = {}

for char in string:
    character_frequency[char] = character_frequency.get(char, 0) + 1



for word in string.split():
    word_frequency[word] = word_frequency.get(word, 0) + 1
```
