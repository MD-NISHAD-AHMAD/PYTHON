```python
# Prime Number Detector 


n = int(input()) 

if n < 2:
    print(f"{n} is not a Prime Number ")
else:
    for i in range(2, n):
        if n % i == 0:
            print(f"{n} is not a prime number")
            break 
    else:
        print(f"{n} is a prime number") 
```

## Pattern 

```python
# Nested Looping 

for outer_loop in range(1, 6):
    print(f"{outer_loop}", end = " --> ")
    for inner_loop in range(11, 14):
        print(f"{inner_loop}", end = " ")
    print() 
```

```python
n = 10

for i in range(n):

    
    for st in range(i + 1):
        print("*", end=" ")

    if i == n - 1:
        for k in range(2 * (n - i - 1)):
            print("*", end=" ")
        for l in range(i):
            print("*", end=" ")
    else:
        for k in range(2 * (n - i - 1) - 1):
            print(" ", end=" ")

        for l in range(i + 1):
            print("*", end=" ")

    print()
```

```python
# reverse pyramid.

n = 8

for i in range(n, 0, -1):

    for j in range(n - i):
        print(" ", end=" ")

    for k in range(2 * i - 1):
        print("*", end=" ")

    print()
```

```python
n = 10


for i in range(n):
    for j in range(n - i - 1):
        print(" ", end=" ")
    for k in range(2 * i + 1):
        print("*", end=" ")
    print()


for i in range(n - 2, -1, -1):
    for j in range(n - i - 1):
        print(" ", end=" ")
    for k in range(2 * i + 1):
        print("*", end=" ")
    print()
```
