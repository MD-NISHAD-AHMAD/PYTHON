## For Loops.

```python
for items in range(1, 11, 1):
    print(items, end= " ")
```

```python
for i in range(10, 1, -1):
    print(i, end=" ")
```

```python
# print even number.

for i in range(2, 11, 2):
    print(i, end=" ")
```

```python
#n print odd number.

for i in range(1, 10, 2):
    print(i, end=" ")
```

```python
# multiplication number.

n = int(input("Enter a number :"))

for i in range(1, 11):
    print(n,*i)
```

```python
n = 5
fact = 1

for i in range(1, n+1):
    fact *= i

print("factorial sum =", fact)    
```
