## if-else clause

```python


age = int(input("Enter the age :"))


if age >= 18: 

    print(f"you can vote since you are {age} years old")

else:
    print(f"you can not vote since you are {age} years old")
```

```python
# Check whether a number is even odd

number = int(input("Enter the number"))

if number % 2 == 0:
    print(f"{number} is even")

else:
    print(f"{number} is odd")
```

```python
n = int(input("Enter the number: "))

if n > 0:
    if n % 2 == 0:
        print(f"{n} - even positive")
    else:
        print(f"{n} - odd positive")

elif n < 0:
    if n % 2 == 0:
        print(f"{n} - even negative")
    else:
        print(f"{n} - odd negative")

else:
    print(f"{n} - zero") 
```

### 4. Leap Year.

```python
year = int(input("Enter a Year"))

if(year % 4 == 0 and year % 100 !=0) or (year % 400 == 0):
    print("Leap year")

else: 
    print("Not a leap year")
```
