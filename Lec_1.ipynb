# 1. Arithmetic Operator

```python
a = 12
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a  ** b}")

print(type(a), id(b))
```

# 2. Comparison Operator

```python
a = 12
b = 5

print(f"a = {a}, b = {b}")
print(f"{a} == {b} : {a == b}")
print(f"{a} != {b} : {a != b}")
print(f"{a} < {b} : {a < b}")
print(f"{a} > {b} : {a > b}")
print( f"{a} <= {b} : {a <= b}")
print(f"{a} >= {b} : {a >= b}")
print(type(a), id(a))
```

# Output Formating

```python
P = 100000
R = 10
T = 60

si = (P * R * T)/(100 * 12)

print(si)
```

# 1. Normal Method

```python
print("the simple interest for" , P, "at", R, '%', 'rate of interest for', T, 'month is', si)
```

## 2. F- String Method (formating string literal)

```python
print(f"The simple Interest for {P} at {R}% rate of interest for {T} month is {si}")
```

# 3. Format Method

```python
print("The simple interest for {} at {}% rate of interes for {} month is {}".format(P, R, T,si))
```
