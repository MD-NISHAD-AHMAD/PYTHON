# Identity Operators

```python
a = 10
b = 10

print(f"{b} == {b} : {a == b}")
print(f"{a} IS {b} : {a is b}")
print(f"memory location of {a} : {id(a)}\nmemory location of {b} : {id(b)}")
```

# Membership Operator


```python
li = [8, 2, 4, 1, 5]

print(f"List : {li}")
print(f"1 in {li} : {1 in li}")
print(f"10 in {li} : {10 in li}")
print(f"100 not in{li} : {100 not in li}")
print(f"100 in li {li} : {100 in li}")
```

# Unary Operator

```python
a = 10 

print(+a)
print(-a)
print(~a)
```
