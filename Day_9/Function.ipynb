```python
# Prime Number 

def isPrime(n : int) -> bool:
    if n < 2:
        return False 
    
    for i in range(2, n):
        if n % i == 0:
            return False 
    else:
        return True 
    

for i in range(1, 101, 1):
    print(f"{i} --> {'Prime' if isPrime(i) == True else 'Not Prime'}")
```

```python
def isLeapYear(year : int) -> None:
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        print(f"{year} is a leap year")
    else:
        print(f"{year} is not a leap year")

return_value = isLeapYear(500) 
print(return_value) 
```

```python
def final(a):
    print(f"Original Variable (Global) : a = {a}")
    # global a 
    a = a + 2 
    print(f"Local Variable : a = {a}") 
a = 10 
print(f"Original Value : a = {a}")
final(a) 
print(f"Updated Value : a = {a}")
```

```python
def f2(a):
    print(f"Global a = {a}, Memory Location : {id(a)}") 
    # local a = global a + 2 
    a = a + 2 
    print(f"Local a = {a}, Memory Location : {id(a)}") 

a = 10 # Global 
print(f"Global a = {a}")  
f2(a) 
print(a) 
```
