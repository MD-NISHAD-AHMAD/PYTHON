## Function

```python
def add(a: int, b: int) -> int:
    return a + b

a = int(input("enter a first number :"))
b = int(input("enter a second number :"))

total = add(a,b)

print(f" the sum of {a} & {b} is {total}")
```

```python
def divide(a: int, b: int) -> float | int:
    if b == 0:
        return "Denominator cannot be zero"
    
    return a / b

# divide(10,0)
divide(10, 5)
```
