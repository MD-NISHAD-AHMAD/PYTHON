# Jump Statement.

## Break ---> stop the iteration.

```python
a = 1

while a < 11:
    print(a, end=" ")

    if a == 5:
        break
    a += 1
```

```python
a = 1
while a <= 10:
    print(a, end= " ")
    if a == 5:
        break

    a += 1
```

## Continue ---> skip the iteration.

```python
a = 1
while a < 11:
    if a % 2 == 1:
        a +=1
        print(a, end=" ")

        a += 1

    else:
        break
```

```python
a = 1
while a < 11:
    if a % 2 == 1:
        a += 1
        continue
    print(a, end=" ")
    a += 1
```

# Infinite loop.

```python
a = 1
while True:
    if a < 6:
        print(a, end=" ")
        a += 1
    else:
        break
```

```python
# print the number from 1 to 10 using infinite loop

a = 1
while True:
    if a > 10:
        break
    print(a, end=" ")
    a += 1
```

```python
# print all the even number from 1 to 10 using infinite loop.

a = 1
while True:
    if a > 10:
        break
    if a % 2 == 0:
        print(a, end=" ")
    a += 1
```

```python
# 2nd method using continue.

a = 1
while True:
    if a > 10:
        break
    if a % 2 != 0:
        a += 1
        continue
    print(a, end=" ")
    a += 1
```

```python
# we can use direct even number.
a = 2
while True:
    if a > 10:
        break
    print(a, end=" ")
    a += 2
```
