# Assignment Operator

```python
a = 10
b = 15

print( f"a = {b}, b = {b}")
a += b
print(f"a = {a}, b = {b}")
a -= b
print(f"a = {a}, b = {b}")
a *= b
print(f"a = {a}, b = {b}")
a /= b
print(f"a = {a}, b = {b}")
a //= b
print(f"a = {a}, b = {b}")
a %= b
print(f"a = {a}, b = {b}")
a **= b
```

# Bitwise Operator

```python
a = 35
b = 30

print(a & b)
print(a | b)
print(a << 1)
print(b >> 2)
print(~a)

```
