# LIST COMPREHENSION.

## Line seperated input.

```python
subjects = ['Marathi', 'Mathematics', 'Science', 'Social Science', 'Hindi'] 
marks_list = [] 
for i in range(len(subjects)):
    marks = int(input(f"Enter the Marks for {subjects[i]} : "))
    marks_list.append(marks) 

print(f"Percentage : {sum(marks_list)/len(marks_list) : 0.2f}%") 
```

## Space seperated input.

```python
marks = input(f"Enter the 5 Subject marks in {subjects} order : ")
print(f"Marks string : {marks}, Data Type : {type(marks)}") 

marks_list = marks.split() 
print(f"Marks List : {marks_list}, Data Type : {type(marks_list)}")

final_list = [] 

for item in marks_list:
    final_list.append(int(item)) 

print(f"Final List : {final_list}, Data Type : {type(final_list)}")
```

```python
from typing import List

# function defintion and declaration

def firstIndex(li : List[int], n : int) -> int:
    for index in range(len(li)):
        if li[index] == n:
            return index
        
    else:
        return -1

    # Taking input from the user.

n = int(input("enter the number :"))
li = [int(item) for item in input("Enter the list elements :").split()]

# function call.

index = firstIndex(li, n)


# output display

if index == -1:
    print(f"{n} not found in {li}")

else:
    print(f"{n} found in {li} at index")
```

### Filter(function, iterable)

#### filters the data based on a particular condition

```python
li = [int(x) for x in range(1, 11, 1)]

print(li)
```

```python
odd_list = list(filter(lambda x : x % 2 != 0, li))
print(odd_list)
```

```python
first_name = ['nishad', 'abdul' ]
last_name = ['ahmad', 'samad']

full_name = list(map(lambda x, y : x + " " + y, first_name, last_name))

print(full_name)
```

```python
li = ["python", "java", "c++", "javascript"]

new_list = list(filter(lambda x : len(x) >= 5, li))

print(f"original list : {li}")
print(f"updated list : {new_list}")
```
