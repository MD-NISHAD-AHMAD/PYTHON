# LIST COMPREHENSION.

## Line seperated input.

```python
subjects = ['Marathi', 'Mathematics', 'Science', 'Social Science', 'Hindi'] 
marks_list = [] 
for i in range(len(subjects)):
    marks = int(input(f"Enter the Marks for {subjects[i]} : "))
    marks_list.append(marks) 

print(f"Percentage : {sum(marks_list)/len(marks_list) : 0.2f}%") 
```

## Space seperated input.

```python
marks = input(f"Enter the 5 Subject marks in {subjects} order : ")
print(f"Marks string : {marks}, Data Type : {type(marks)}") 

marks_list = marks.split() 
print(f"Marks List : {marks_list}, Data Type : {type(marks_list)}")

final_list = [] 

for item in marks_list:
    final_list.append(int(item)) 

print(f"Final List : {final_list}, Data Type : {type(final_list)}")
```

```python
# get all the even numbers from 1 to 100 in a list 

# Extended Version 

li = [] 

for item in range(1, 101, 1):
    if item % 2 == 0:
        li.append(item) 

print(li) 
```

```python
new_list = [item for item in range(1, 101, 1) if item % 2 == 0]
print(new_list)
```

```python
# Save all numbers from 1 to 30 in a list 

newli = [item for item in range(1, 31, 1) if item % 2 == 1]
print(newli)
```

```python
ans = [item for item in range(1, 101, 1) if item % 5 == 0]
print(ans)
```

```python
# create a list such  that it has capital alphabets 

capital_alphabets = [chr(i) for i in range(ord("A"), ord("Z") + 1, 1)]
print(capital_alphabets)

small_alphabets = [chr(i) for i in range(ord("a"), ord("z") + 1, 1)]
print(small_alphabets)
```
