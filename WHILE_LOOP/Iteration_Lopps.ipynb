#### print a number from 1 to 10

```python
start = int(input("enter a start number :"))
end = int(input("enter a end number :"))

count = start

while count < end +1:

    print(count, end =" ")

    count = count +1
```

### odd or even number

```python
start = 1
stop = 10

while start <= stop:
  if start % 2 == 0:
    print(start, end = " ")
    start += 1
  else:
    start += 1
```

```python
# sum of n natural number.

start = 1

stop = int(input("enter a number :"))

sum = 0

while start <= stop:
    
    sum = sum + start
    

    start += 1

else:
    print(sum)

   
```

```python
# print reverse a number.

n = int(input("enter a number :"))

while n != 0:

    rem = n % 10

    print(f"{rem}")
    n = n // 10

```

```python
# sum of add digits number.

n = int(input("enter a number :"))

total = 0

ref = n

while n != 0:
    total += n % 10

    n = n // 10

else:
    print(f"sum of digits {ref} is {total}")
```
