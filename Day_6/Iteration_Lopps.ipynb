#### print a number from 1 to 10

```python
start = int(input("enter a start number :"))
end = int(input("enter a end number :"))

count = start

while count < end +1:

    print(count, end =" ")

    count = count +1
```

### odd or even number

```python
start = 1
stop = 10

while start <= stop:
  if start % 2 == 0:
    print(start, end = " ")
    start += 1
  else:
    start += 1
```
